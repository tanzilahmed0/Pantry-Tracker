## Problem 3 (10 points)
-  Add a new column `Age_group` to the dataframe from **Problem 1** with categorical values according to the following rules.
-  Show the resulting dataframe.

| Age | Age_group |
|---|---|
| 15-24 | Youth |
| 25-64 | Adult |
| 65+ | Elderly |

```python
# Problem 3 Answer:
def age_g(x):
    if '15' in x:
        return 'Youth'
    if '25' in x:
        return 'Adult'
    else:
        return 'Elderly'

age_group = df['Age'].map(age_g)
df['Age_group'] = age_group

df.head()
```
